```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error
import pandas as pd
```

```python
#you are telling Python to ignore all warning messages, 
#that would typically be displayed during the execution of your code.
import warnings
warnings.filterwarnings("ignore") 
```

```python
X = 2 * np.random.rand(1000, 1)
y = 4 + 3 * X 
```

```python
GD_model = SGDRegressor(max_iter=1, eta0=1, warm_start=True)
```

```python
GD_model
```

```python
out = []
for i in range(20):
    GD_model.fit(X, y,)
    out.append([
                i, GD_model.coef_[0], GD_model.intercept_[0], mean_squared_error(y, GD_model.predict(X))
                ])
    df = pd.DataFrame(out, columns=["itteration", "m", "b", "MSE"])
    print(out[-1])
    input()
    
    
    
    
    
    
    
    
    
```
