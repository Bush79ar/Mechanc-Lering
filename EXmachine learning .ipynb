```python

# Sample data for binary classification
X = np.array([2, 3, 5, 7, 8, 9]).reshape(-1, 1)  # Hours of sunshine
y = np.array([0, 0, 1, 1, 1, 1])  # Binary class labels (0 or 1)

# Use Linear Regression for classification
from sklearn.linear_model import LinearRegression

# Create and train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the data and the decision boundary (regression line)
plt.scatter(X, y, color='b')
plt.plot(X, y_pred, color='r')
plt.xlabel('Hours of Sunshine')
plt.ylabel('Class Label (0 or 1)')
plt.title('Linear Regression for Classification')
plt.show()
```

```python
ols = LinearRegression()
ols.fit(X, y)
```

```python

# Create and train the Logistic Regression model
model = LogisticRegression()
model.fit(X, y)

# Make predictions
y_pred_proba = model.predict_proba(X)[:, 1]  # Predict probabilities of positive class (class 1)
y_pred = model.predict(X)  # Predict class labels (0 or 1)

# Plot the data and the decision boundary (logistic function)
plt.scatter(X, y, color='b')
plt.plot(X, y_pred_proba, color='r', label='Logistic Regression')
plt.xlabel('Hours of Sunshine')
plt.ylabel('Class Label (0 or 1)')
plt.title('Logistic Regression for Binary Classification')
plt.legend()
plt.show()
```

```python
clf = LogisticRegression(C=1e5)
clf.fit(X, y)
```

```python
# and plot the result
plt.figure(1, figsize=(4, 3))
plt.clf()
plt.scatter(X.ravel(), y, label="example data", color="black", zorder=20)
X_test = np.linspace(-5, 10, 300)

# loss = expit(X_test * clf.coef_ + clf.intercept_).ravel()
loss = clf.predict_proba(X_test.reshape((-1, 1)))[:,1]

plt.plot(X_test, loss, label="Logistic Regression Model", color="red", linewidth=3)



plt.plot(
    X_test,
    ols.coef_ * X_test + ols.intercept_,
    label="Linear Regression Model",
    linewidth=1,
)
plt.axhline(0.5, color=".5")

plt.ylabel("y")
plt.xlabel("X")
plt.xticks(range(-5, 10))
plt.yticks([0, 0.5, 1])
plt.ylim(-0.25, 1.25)
plt.xlim(-4, 10)
plt.legend(
    loc="lower right",
    fontsize="small",
)
plt.tight_layout()
plt.show()
```
